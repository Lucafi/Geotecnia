# Fundações Rasas - Recalques em areias

O recalque em areias apresenta o módulo de deformabilidade variável com a profundidade, ainda que se garanta a homogeneidade granular, mineralógica e de compacidade. Esse é um fato descrito em diversas referências que tratam sobre o assunto (Anderson, Townsend, & Rahelison, 2007;  Fonseca, 2001; Schmertmann, 1978, D'Apollonia & D'Apollonia, 1970). 

Embora, o meio elástico não homogêneo, característico das areias e ainda que o confinamento aumente o módulo de deformabilidade, há vários métodos para estimativa de recalque para fundações rasas reunidos no escopo conhecido como Meio Elástico Homogêneo (**MEH**). Isso, parece à primeira vista contraditório, mas se dividirmos em camadas pouco espessas, poderemos considerar valor constante para o módulo de deformabilidade $E_s $.

Segundo Cintra et al. (2014), a divisão em camadas é essencial para o método de previsão de recalques de fundações rasas ema reias. O método Schmertmann, divide o solo abaixo do contato da fundação em camadas, ancorando na Teoria da Elasticidade para tomar uma relação da variabilidade da deformabilidade com a profundidade da camada, tanto na versão de 1970 quanto na de 1978.

## Método Schmertmann (1970)

Um carregamento uniforme ($\sigma$) solicita a superfície homogênea de um espaço semielástico (centro do carregamento), isotrópico, homogêneo com módulo de elasticidade ($E_s$), deformação vertical ($\varepsilon_z$), à profundidade $z$ pode ser expressa pela Eq. 1:


$$\varepsilon_z=\frac{\sigma}{E_s}I_z \ \ \ \ \ (1)$$

em que $I_z$ é o fator de influência na deformação do solo atuado pelo carregamento.

### Observações de Schmertmann (1970)
Através de análises teóricas, modelagem, e simulações pelo método dos elementos finitos, o autor pesquisou a variação da deformação vertical ao longo da profundidade, em solos arenosos homogêneos, sob sapatas rígidas.

Observou que:
* a deformação máxima não ocorre no contato com a base da fundação rasa, mas a uma profundidade em torno de $z = \frac{B}{2}$;
* a partir da  profundidade $\frac{B}{2}$, as deformações diminuem gradualmente e podem ser desprezadas depois de $z = 2B$, em que $B$ é a largura da fundação rasa;
* o autor propõe uma distribuição aproximada do fator de influência na deformação para o cálculo de recalque de sapatas rígidas em areia, apresentada no diagrama triangular da Fig. 1.

O Diagrama da Figura 1 representa o bulbo de recalques devido à pressão de contato, igual a $z=2B$, que coincide com o bulbo de tensões. Contudo, observa-se que para um blbo de tensão que esgote 90% do recalque possível aprofundidade de influência deveria ser $z=6B$ (Cintra et al., 2014).

**Figura 1.** Diagrama do fator de influência de acordo com a profundidade

### Embutimento da fundação

Cintra et al. (2014), alertam que o embutimento, por exemplo em sapata rasa, pode diminir o recalque em até 50%, comprovado pelo método de camada finita. Schmertmann (1970), faz uma correção do recalque pela formulação de Eq.2:

$$C_1 = 1 - 0,5\left(\frac{q}{\sigma^*}\right)\geq 0,5 \ \ \ \ (2)$$                                 

em que,
q é atensão vertical efetiva na cota de assentamento da fundação (sobrecarga).

$\sigma^*$ é a tensão líquida dada pela Eq. 3

$$ \sigma^* =\sigma -q \ \ \ \ \ (3)$$

A tensão líquida desconta da tensão aplicada o valor correpondente à sobrecarga e que não deve gerar recalque, uma vez que corresponde à reposição do alívio de tensões originado das escavações. No caso de sapatas essa diferença pode até ser negligenciada, porém em tubulões a diferença é notável.


### Efeito do tempo

O recalque imediato em areias é perceptível, às vezes visualmente. Há porém recalques que se desenvolvem ao longo do tempo ($t$) como se fossem os recalques devido à compressão secundária na argila. Schmertmann (1970) adotou a Eq.4 para essa correção de recalques:

$$C_2 = 1+0,2 log \left(\frac{t}{0,1}\right) \ \ \ \ \ (4)$$

com $t$ em anos.

Observa-se que em 1 ano o acréscimo de recalques corresponde a $20\%$; a 10 anos a $40\%$ e a 100 anos $60\%$. Se o interess é o recalque imadiato, então $C_2=1$.

### Formulação

O recalque de sapatas rígidas em areia (ρd) é dado pelo somatório dos recalques de n subcamadas consideradas homogêneas, na profundidade de 0 a 2B, incluindo os efeitos do embutimento e do tempo (Cintra et al, 2014), conforme Schmertamnn (1970), o recalque é dado pela a Eq. 5:

$$ \rho_d = C_1 C_2 \sigma^* \sum_{i=1}^{n} \ \ \ \ \ (5)$$

em que:
* $Iz$ é o fator de influência na deformação à meia-altura da i-ésima camada;
* $E_s$ é o módulo de deformabilidade da i-ésima camada;
*$\Delta Z$ é a espessura da i-ésima camada.

## Método Schmertmann (1978)

Em 1978 Schmertmann apresentou uma melhoria do seu método com a separação as fundações corridas das quadradas. Dois diagramas são propostos como se observa na Figura 2. Nota-se que o bulbo de recalques é maior para as sapatas corridas, os valores iniciais de $I_z$ são diferentes de zero e $I_{zmáx}$, Eq. 8, não é fixo e difere entre sapata quadrada e corrida.

$$I_{zmáx}=0,5+0,1 \sqrt{\frac{\sigma^*}{\sigma_v}}\ \ \ \ \ (8) $$

em que $\sigma_v$ é a tensão vertical efetiva  na profundidade correspondente. 

Esse método foi proposto para sapatas, mas pode ser aplicado a tubulões, nesse caso $\frac{L}{B}=1$.

**Figura 2.** Fator de influência na deformação vertical (Schmertmann, 1978).



## Método da Camada Finita - Sapata Fictícia
A camada de solo, em muitas situações, tem espessura finita e definida para o impenetrável a SPT, por exemplo. Essa camada ou camadas (podem ser várias até atingir o topo rochoso) pode ter estimada, a sua deformabilidade, considerando a seguinte condição de contorno: um meio elástico homogêneo sobrejacente a uma camada indeslocável para as tensões comumentes da engenharia. A solução dessa condição, expressa na Eq. 9, foi feita por Janbu et al.(1956), considerando deformações de volume constante ($V=0,5$), uma referência direta ao comportamento de argilas saturadas não drenadas e em acréscimo a adaptação da teoria da elasticidade para a obtenção de recalque médio de sapatas flexíveis (Cintra et al., 2014). 

$$ \rho_i=\mu_0 \mu_1 \frac{\sigma B}{E_s}\ \ \ \ \ (9)$$

onde $\mu_0$ e $\mu_1$ são **fatores de influência** do embutimento da sapata e da espessura da camada de solo.

Observa-se, na Figura 3,  que o fator $\mu_0$ varia de 0,5 a 1,0, ou seja, o recalque pode ser reduzido em até 50%, devido a esse fator.


**Figura 3.** Faores de influência para embutimento e espessura da camda de solo

## Solução
**a) Determinação de $q$**

**Segunda camada: -4,40  a -5,50 m**

*Tabela 4*

$h_3 =3,50 \ m$

```python
h1=4.4;h2=1.1;h3=3.50
gamma1=13;gamma2=16;gamma3=16
q=((h1*gamma1)+(h2*gamma2)+(h3*gamma3))
print(round(q,2),"kPa")
```

**b) Determinação $\sigma^*$, $C_1$ e $C_2$**

```python
sigma=400
sigmaL=sigma-q
C1=1-(q/sigmaL)
C2=1

print("A tensão líqida é", sigmaL,"kPa")
print(" O valor de C1 é: ", round(C1,2))
print("O recalque é imediato, então C2 é:",C2)
```

### Cálculo de uma base "quadrada" equivalente
Calcula-se um $L^2$ equivalente à área da base do tubulão $S$.

Além disso como é uma base quadrada $L=B$.

```python
Dt= 3.0 # Diâmetro do tubulão
S=(np.pi*Dt**2)/4 # Área do tubulão
L=np.sqrt(S)
print("Sapata equivalente com L = ",round(L,1), "metros" )
```

### Determinação de $I_{zmáx}$.
Determina-se a $\sigma_v$ na cota de apoio do tubulão:

$\sigma_v=q+\frac{B}{2}\times \gamma$

No caso, $\gamma=\gamma_3$.

```python
B=round(L,1)
sigmav=q+(B/2)*gamma3
print("A tensão vertical na cota de Izmax é igual a:", sigmav)
Izmax=0.5+0.1*(np.sqrt(sigmaL/sigmav))
Izmax=round(Izmax,2)
print("Izmax é igual a: ", Izmax)
```
