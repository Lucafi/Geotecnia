# Fundaes Rasas - Recalques em areias

O recalque em areias apresenta o mdulo de deformabilidade varivel com a profundidade, ainda que se garanta a homogeneidade granular, mineralgica e de compacidade. Esse  um fato descrito em diversas referncias que tratam sobre o assunto (Anderson, Townsend, & Rahelison, 2007;  Fonseca, 2001; Schmertmann, 1978, D'Apollonia & D'Apollonia, 1970). 

Embora, o meio elstico no homogneo, caracterstico das areias e ainda que o confinamento aumente o mdulo de deformabilidade, h vrios mtodos para estimativa de recalque para fundaes rasas reunidos no escopo conhecido como Meio Elstico Homogneo (**MEH**). Isso, parece  primeira vista contraditrio, mas se dividirmos em camadas pouco espessas, poderemos considerar valor constante para o mdulo de deformabilidade $E_s $.

Segundo Cintra et al. (2014), a diviso em camadas  essencial para o mtodo de previso de recalques de fundaes rasas ema reias. O mtodo Schmertmann, divide o solo abaixo do contato da fundao em camadas, ancorando na Teoria da Elasticidade para tomar uma relao da variabilidade da deformabilidade com a profundidade da camada, tanto na verso de 1970 quanto na de 1978.

## Mtodo Schmertmann (1970)

Um carregamento uniforme ($\sigma$) solicita a superfcie homognea de um espao semielstico (centro do carregamento), isotrpico, homogneo com mdulo de elasticidade ($E_s$), deformao vertical ($\varepsilon_z$),  profundidade $z$ pode ser expressa pela Eq. 1:


$$\varepsilon_z=\frac{\sigma}{E_s}I_z \ \ \ \ \ (1)$$

em que $I_z$  o fator de influncia na deformao do solo atuado pelo carregamento.

### Observaes de Schmertmann (1970)
Atravs de anlises tericas, modelagem, e simulaes pelo mtodo dos elementos finitos, o autor pesquisou a variao da deformao vertical ao longo da profundidade, em solos arenosos homogneos, sob sapatas rgidas.

Observou que:
* a deformao mxima no ocorre no contato com a base da fundao rasa, mas a uma profundidade em torno de $z = \frac{B}{2}$;
* a partir da  profundidade $\frac{B}{2}$, as deformaes diminuem gradualmente e podem ser desprezadas depois de $z = 2B$, em que $B$  a largura da fundao rasa;
* o autor prope uma distribuio aproximada do fator de influncia na deformao para o clculo de recalque de sapatas rgidas em areia, apresentada no diagrama triangular da Fig. 1.

O Diagrama da Figura 1 representa o bulbo de recalques devido  presso de contato, igual a $z=2B$, que coincide com o bulbo de tenses. Contudo, observa-se que para um blbo de tenso que esgote 90% do recalque possvel aprofundidade de influncia deveria ser $z=6B$ (Cintra et al., 2014).

**Figura 1.** Diagrama do fator de influncia de acordo com a profundidade

### Embutimento da fundao

Cintra et al. (2014), alertam que o embutimento, por exemplo em sapata rasa, pode diminir o recalque em at 50%, comprovado pelo mtodo de camada finita. Schmertmann (1970), faz uma correo do recalque pela formulao de Eq.2:

$$C_1 = 1 - 0,5\left(\frac{q}{\sigma^*}\right)\geq 0,5 \ \ \ \ (2)$$                                 

em que,
q  atenso vertical efetiva na cota de assentamento da fundao (sobrecarga).

$\sigma^*$  a tenso lquida dada pela Eq. 3

$$ \sigma^* =\sigma -q \ \ \ \ \ (3)$$

A tenso lquida desconta da tenso aplicada o valor correpondente  sobrecarga e que no deve gerar recalque, uma vez que corresponde  reposio do alvio de tenses originado das escavaes. No caso de sapatas essa diferena pode at ser negligenciada, porm em tubules a diferena  notvel.


### Efeito do tempo

O recalque imediato em areias  perceptvel, s vezes visualmente. H porm recalques que se desenvolvem ao longo do tempo ($t$) como se fossem os recalques devido  compresso secundria na argila. Schmertmann (1970) adotou a Eq.4 para essa correo de recalques:

$$C_2 = 1+0,2 log \left(\frac{t}{0,1}\right) \ \ \ \ \ (4)$$

com $t$ em anos.

Observa-se que em 1 ano o acrscimo de recalques corresponde a $20\%$; a 10 anos a $40\%$ e a 100 anos $60\%$. Se o interess  o recalque imadiato, ento $C_2=1$.

### Formulao

O recalque de sapatas rgidas em areia (d)  dado pelo somatrio dos recalques de n subcamadas consideradas homogneas, na profundidade de 0 a 2B, incluindo os efeitos do embutimento e do tempo (Cintra et al, 2014), conforme Schmertamnn (1970), o recalque  dado pela a Eq. 5:

$$ \rho_d = C_1 C_2 \sigma^* \sum_{i=1}^{n} \ \ \ \ \ (5)$$

em que:
* $Iz$  o fator de influncia na deformao  meia-altura da i-sima camada;
* $E_s$  o mdulo de deformabilidade da i-sima camada;
*$\Delta Z$  a espessura da i-sima camada.

## Mtodo Schmertmann (1978)

Em 1978 Schmertmann apresentou uma melhoria do seu mtodo com a separao as fundaes corridas das quadradas. Dois diagramas so propostos como se observa na Figura 2. Nota-se que o bulbo de recalques  maior para as sapatas corridas, os valores iniciais de $I_z$ so diferentes de zero e $I_{zmx}$, Eq. 8, no  fixo e difere entre sapata quadrada e corrida.

$$I_{zmx}=0,5+0,1 \sqrt{\frac{\sigma^*}{\sigma_v}}\ \ \ \ \ (8) $$

em que $\sigma_v$  a tenso vertical efetiva  na profundidade correspondente. 

Esse mtodo foi proposto para sapatas, mas pode ser aplicado a tubules, nesse caso $\frac{L}{B}=1$.

**Figura 2.** Fator de influncia na deformao vertical (Schmertmann, 1978).



## Mtodo da Camada Finita - Sapata Fictcia
A camada de solo, em muitas situaes, tem espessura finita e definida para o impenetrvel a SPT, por exemplo. Essa camada ou camadas (podem ser vrias at atingir o topo rochoso) pode ter estimada, a sua deformabilidade, considerando a seguinte condio de contorno: um meio elstico homogneo sobrejacente a uma camada indeslocvel para as tenses comumentes da engenharia. A soluo dessa condio, expressa na Eq. 9, foi feita por Janbu et al.(1956), considerando deformaes de volume constante ($V=0,5$), uma referncia direta ao comportamento de argilas saturadas no drenadas e em acrscimo a adaptao da teoria da elasticidade para a obteno de recalque mdio de sapatas flexveis (Cintra et al., 2014). 

$$ \rho_i=\mu_0 \mu_1 \frac{\sigma B}{E_s}\ \ \ \ \ (9)$$

onde $\mu_0$ e $\mu_1$ so **fatores de influncia** do embutimento da sapata e da espessura da camada de solo.

Observa-se, na Figura 3,  que o fator $\mu_0$ varia de 0,5 a 1,0, ou seja, o recalque pode ser reduzido em at 50%, devido a esse fator.


**Figura 3.** Faores de influncia para embutimento e espessura da camda de solo

## Soluo

```python
h1=4.4;h2=1.1;h3=3.50
gamma1=13;gamma2=16;gamma3=16
q=((h1*gamma1)+(h2*gamma2)+(h3*gamma3))
print(round(q,2),"kPa")
```

**b) Determinao $\sigma^*$, $C_1$ e $C_2$**

```python
sigma=400
sigmaL=sigma-q
C1=1-(q/sigmaL)
C2=1

print("A tenso lqida ", sigmaL,"kPa")
print(" O valor de C1 : ", round(C1,2))
print("O recalque  imediato, ento C2 :",C2)
```

### Clculo de uma base "quadrada" equivalente
Calcula-se um $L^2$ equivalente  rea da base do tubulo $S$.

Alm disso como  uma base quadrada $L=B$.

```python
Dt= 3.0 # Dimetro do tubulo
S=(np.pi*Dt**2)/4 # rea do tubulo
L=np.sqrt(S)
print("Sapata equivalente com L = ",round(L,1), "metros" )
```

### Determinao de $I_{zmx}$.
Determina-se a $\sigma_v$ na cota de apoio do tubulo:

$\sigma_v=q+\frac{B}{2}\times \gamma$

No caso, $\gamma=\gamma_3$.

```python
B=round(L,1)
sigmav=q+(B/2)*gamma3
print("A tenso vertical na cota de Izmax  igual a:", sigmav)
Izmax=0.5+0.1*(np.sqrt(sigmaL/sigmav))
Izmax=round(Izmax,2)
print("Izmax  igual a: ", Izmax)
```
